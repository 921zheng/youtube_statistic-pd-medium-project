### length is shorter than 30 min is short video; else long video

```python
#compare likes between short and long videos
df['type']=np.where(df['length']>=30, 'long', 'short')
df
```

```python
#  compare count of likes on video length
df[df['type']=='long']['likes'].sum()
```

```python
df[df['type']=='short']['likes'].sum()
```

```python
#compare count of dislikes on video length
df[df['type']=='long']['dislikes'].sum()
```

```python
df[df['type']=='short']['dislikes'].sum()
```

```python
#compare count of comment on video length
df[df['type']=='long']['comment_count'].sum()
```

```python
df[df['type']=='short']['comment_count'].sum()
```
